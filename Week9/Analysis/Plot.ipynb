```python
import MDAnalysis as md
import numpy as np
import matplotlib.pyplot as plt
from MDAnalysis.analysis import msd

# List of universes
universes = [
    md.Universe("../Data/equil/kalj_T0.45_n360_v300_100000000_1.lammpstrj", "../Data/prod/kalj_T0.45_n360_v300_prod_100000000_1.unwrap.dcd", topology_format="LAMMPSDUMP"),
    md.Universe("../Data/equil/kalj_T0.475_n360_v300_30000000_1.lammpstrj", "../Data/prod/kalj_T0.475_n360_v300_prod_30000000_1.unwrap.dcd", topology_format="LAMMPSDUMP"),
    md.Universe("../Data/equil/kalj_T0.5_n360_v300_10000000_1.lammpstrj", "../Data/prod/kalj_T0.5_n360_v300_prod_10000000_1.unwrap.dcd", topology_format="LAMMPSDUMP"),
    md.Universe("../Data/equil/kalj_T0.55_n360_v300_3000000_1.lammpstrj", "../Data/prod/kalj_T0.55_n360_v300_prod_3000000_1.unwrap.dcd", topology_format="LAMMPSDUMP"),
    md.Universe("../Data/equil/kalj_T0.6_n360_v300_1000000_1.lammpstrj", "../Data/prod/kalj_T0.6_n360_v300_prod_1000000_1.unwrap.dcd", topology_format="LAMMPSDUMP"),
    md.Universe("../Data/equil/kalj_T0.65_n360_v300_350000_1.lammpstrj", "../Data/prod/kalj_T0.65_n360_v300_prod_350000_1.unwrap.dcd", topology_format="LAMMPSDUMP"),
    md.Universe("../Data/equil/kalj_T0.7_n360_v300_180000_1.lammpstrj", "../Data/prod/kalj_T0.7_n360_v300_prod_180000_1.unwrap.dcd", topology_format="LAMMPSDUMP"),
    md.Universe("../Data/equil/kalj_T0.8_n360_v300_150000_1.lammpstrj", "../Data/prod/kalj_T0.8_n360_v300_prod_150000_1.unwrap.dcd", topology_format="LAMMPSDUMP"),
    md.Universe("../Data/equil/kalj_T0.9_n360_v300_150000_1.lammpstrj", "../Data/prod/kalj_T0.9_n360_v300_prod_150000_1.unwrap.dcd", topology_format="LAMMPSDUMP"),
    md.Universe("../Data/equil/kalj_T1_n360_v300_50000_1.lammpstrj", "../Data/prod/kalj_T1_n360_v300_prod_50000_1.unwrap.dcd", topology_format="LAMMPSDUMP"),
    md.Universe("../Data/equil/kalj_T1.5_n360_v300_10000_1.lammpstrj", "../Data/prod/kalj_T1.5_n360_v300_prod_10000_1.unwrap.dcd", topology_format="LAMMPSDUMP")
]

def compute_msd(atom_group, prd_data):
    type1 = prd_data.select_atoms(atom_group)
    msd1 = msd.EinsteinMSD(type1).run()
    return msd1

# Define the particle type
type1_atoms = "type 1"

# Compute MSD for type 1 particles at each temperature
temperatures = [0.475, 0.5, 0.55, 0.6, 0.65, 0.7, 0.8, 0.9, 1.0, 1.5]  # Add your temperatures here
time_msd_1={}
plt.figure(figsize=(10, 6))
for i, temp in enumerate(temperatures):
    # Assuming each temperature corresponds to one universe, you can use indexing
    current_universe = universes[i]

    # Compute RDF for the specific temperature
    msd_dict = compute_msd(type1_atoms, current_universe)

    # Plot RDF for the current temperature
    plt.loglog(msd_dict.times, msd_dict.run().results['msds_by_particle'].mean(axis=1), label=f'Temperature {temp}')
    # Find the time where MSD = 1
    time_msd_1[i] = msd_dict.times[np.argmax(msd_dict.run().results['msds_by_particle'].mean(axis=1) >= 1)]

plt.title('Mean-Squared Displacement for Type 1 Particles')
plt.xlabel('Time')
plt.ylabel('MSD')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.savefig('MSDtype1.png')
plt.show()
```

```python
# Convert temperatures and time_msd_1 to numpy arrays
temperatures_np = np.array(temperatures)
time_msd_1_np = np.array(list(time_msd_1.values()))

# Plotting log-log scale
plt.loglog(1 / temperatures_np, time_msd_1_np, marker='o', linestyle='-', color='b', label='Time where MSD=1')

plt.title('Time where MSD=1 vs 1/Temperature')
plt.xlabel('1/Temperature')
plt.ylabel('Time where MSD=1')
plt.legend()
plt.savefig('MSDtype1.png')
plt.show()
```
