```python
import numpy as np
import pandas as pd
import os
import matplotlib as mpl
import matplotlib.pyplot as plt


plt.figure(figsize=(8,5),dpi=600)
font = {'family' : 'serif','weight' : 'bold','size' : 22}
font1 = {'family' : 'serif','weight' : 'bold','size' : 26}
font2 = {'family' : 'serif','weight' : 'bold','size' : 30}
plt.rcParams["axes.linewidth"] = 2.0
plt.rcParams["axes.labelsize"] = 26
#plt.rcParams["axes.titlepad"] = 14
#plt.suptitle('C60 system(108 molecule) dt=0.0005 ps',fontsize=30,weight='semibold',y=0.92)'

plt.subplot(1,1,1)

filename2 = '../1_NPT/T295/energymin.xvg'
X,Y2 = [],[]
with open(filename2, 'r') as f:#1
    lines = f.readlines()#2
    for line in lines:#3
        value = [float(s) for s in line.split()]#4
        X.append(value[0])#5
        Y2.append(value[1])
plt.plot(X,Y2, color = 'black',linewidth=3.0, linestyle= '-',label ='T = 295 K',ms=10)   
plt.rcParams.update({'font.size': 20})
#plt.text(0.01,16.9,"(a) $\mathbf{ChCl:Eth}$ ",fontsize=38,weight='semibold',alpha=1)
plt.legend(frameon=False,loc="upper right",prop=font)
#plt.text(0.5,0.3,"(a) Mt=0.001",fontsize=12,alpha=1)  medium
plt.tick_params(direction='in',labelsize=20, length=6, width=3,top=True,right=True)
plt.xlabel(r't/ps',font1)
plt.ylabel(r'potential energy',font1)

plt.subplots_adjust(wspace=0,hspace=0,left=0.15)
plt.savefig('../Figures/min1.png', dpi=1000, bbox_inches='tight',facecolor='white')
```
