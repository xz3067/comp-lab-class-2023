```python
import mdtraj as md
import numpy as np
import matplotlib.pyplot as plt

# Load AlphaFold predictions and the crystal structure
alphafold_traj = md.load("test_ce6c6_unrelaxed_rank_001_alphafold2_ptm_model_5_seed_000.pdb")
crystal_traj = md.load("1lys.pdb")

# Align AlphaFold predictions to the crystal structure based on alpha-carbons
alphafold_traj_aligned = alphafold_traj.superpose(crystal_traj, atom_indices=alphafold_traj.top.select_atom_indices("alpha"))

# You can also align based on the full backbone and side-chains if needed
# alphafold_traj_aligned = alphafold_traj.superpose(crystal_traj, atom_indices=alphafold_traj.top.select_atom_indices("backbone and not hydrogen"))

# Calculate RMSD for each frame based on alpha-carbons
rmsd_alpha = md.rmsd(alphafold_traj_aligned, crystal_traj, atom_indices=alphafold_traj.top.select_atom_indices("alpha"))

# Calculate RMSD for each frame based on full backbone and side-chains if needed
# rmsd_full = md.rmsd(alphafold_traj_aligned, crystal_traj, atom_indices=alphafold_traj.top.select_atom_indices("backbone and not hydrogen"))
# Create a bar plot for RMSD based on alpha-carbons
plt.bar(range(len(rmsd_alpha)), rmsd_alpha, label="Alpha-Carbons")
plt.xlabel("Frame")
plt.ylabel("RMSD (Å)")
plt.title("RMSD of AlphaFold Predictions to Crystal Structure")
plt.legend()
plt.savefig('RMSD_1.png')
plt.tight_layout()
plt.show()


```
